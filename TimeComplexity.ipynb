```python
def binar(arr,target):
  low,high = 0, len(arr)-1

  while low <= high:
    mid = low + high // 2
    if arr[mid] == target:
      return mid
    elif arr[mid] < target:
      low = mid + 1
    else:
      high = mid - 1
  return -1

arr = [1,23,34,45,55,67,71,78,80,90,91,95,99,100,101,120,140,150,165]
target = 71
result = binar(arr, target)
#print(result)

if result != -1:
  print(f"found , {target} at Index {result}")
else:
  print("not found")
```
