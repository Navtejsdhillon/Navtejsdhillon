The Tower of Hanoi is a classic puzzle that involves three rods and a number of disks of different sizes, which can slide onto any rod. The puzzle starts with the disks in a neat stack in ascending order of size on one rod, the smallest at the top (i.e., smallest disk on top, larger disks below it). The objective is to move the entire stack to another rod, obeying the following rules:

Only one disk can be moved at a time.
Each move consists of taking the upper disk from one of the stacks and placing it on top of another stack or on an empty rod.
No disk may be placed on top of a smaller disk.
The goal of the Tower of Hanoi puzzle is to move all the disks from the source rod to the target rod using the auxiliary rod as needed.

Recursion is often used to solve the Tower of Hanoi problem efficiently by breaking it down into smaller sub-problems. In the recursive solution, you can imagine that you are solving the Tower of Hanoi problem for a smaller number of disks, and then use that solution to solve the problem for larger numbers of disks.

Sum of natural numbers upto n using recursion

```python
def sum_num(n):
  if n == 1:
    return 1
  else:
    return n + sum_num(n - 1)

num4 = int(input("Number: "))
if num4 <=0:
  print("invalid")
else:
  print(num4,sum_num(num4))
```

Factorial

```python
def factorial(n):
  if n==0:
    return 1
  else:
    return n*factorial(n-1)

factorial(10)
```
