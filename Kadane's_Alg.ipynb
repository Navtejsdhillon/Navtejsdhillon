```python
#Kadane's Algorithm
a = [1,2,3,-2,5]
n = 0
l = []
max = 0
for i in range(len(a)):
  n = n + a[i]
  l.append(n)
  l.sort(reverse=True)
print("highest sum ",l[0])

```

```python
a = [2,1,0,3,4,0]
n = 0
l = []
for i in range(len(a)):
  for j in range(i,len(a)):
    if a[i] == a[j]:
      if a[i] == n and a[j]==n:
        l.append((a[i],a[j]))
      n += 1
print(l)
```

```python
a = [2, 1, 0, 3, 4, 0]
n = 0
l = []

for i in a:
    if i == n:
        l.append(i)
        n += 1
    else:
      if len(l) > 1:
            l.clear()
            l.append(i)
            n = 0

print(l)
```
