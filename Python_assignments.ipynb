Armstrong Award Easy In a maths competition, the Armstrong award would be presented to the one who would first tell the Armstrong
number among all given numbers. Thus, write a program for Sam in Python to help him won the award.
What is Armstrong Number?
It is a number which is equal to the sum of cube of its digits. For eg: 153, 370 etc.
Let's take 153 for an example
First calculate the cube of its each digits
1^3 = (1 * 1 * 1) = 1
5^3 = (5 * 5 * 5) = 125
3^3= (3 * 3 * 3) = 27
Now add the cube
1+125+27 = 153
It means 153 is an Armstrong Number.
Example 1:
Input:
153
Output:
Number is Armstrong
Example 2:
Input:
450
Output:
Number is not Armstrong

```python
def armstrong_number(num):
  num1 = str(num)
  num_d = len(num1)
  sum = 0
  for x in num1:
    digit = int(x)
    sum = sum+digit**num_d
  return sum == num
num = 153
if armstrong_number(num):
  print("yes")
else:
  print("No")

armstrong_number(num)
```

Integer with 3 primes:
Write a Python program to find all integers <= 1000 that are the product of exactly three primes.
Each integer should represent as the list of its three prime factors.
Example 1:
Input:
10
Output:
[[2, 2, 2]]

```python
import sympy
primes = list(sympy.primerange(1,1000))

def three_prime(number):
  list_prime = []

  for i in range(len(primes)):
    for j in range(i + 1, len(primes)):
      for k in range(j + 1, len(primes)):
        prod = primes[i]*primes[j]*primes[k]
        if prod <= 1000:
          list_prime.append([primes[i],primes[j],primes[k]])
    print(list_prime)


three_prime(number)

```

```python
def count_palindromic_substrings(s):
    def is_palindrome(sub):
        return sub == sub[::-1]

    def count_palindromic_substrings_helper(s, start, end):
        if start >= end:
            return 0
        if is_palindrome(s[start:end + 1]):
            return 1 + count_palindromic_substrings_helper(s, start + 1, end) + count_palindromic_substrings_helper(s, start, end - 1)
        else:
            return count_palindromic_substrings_helper(s, start + 1, end) + count_palindromic_substrings_helper(s, start, end - 1) - count_palindromic_substrings_helper(s, start + 1, end - 1)

    return count_palindromic_substrings_helper(s, 0, len(s) - 1)

# Example 1
input_str1 = "abcab"
output1 = count_palindromic_substrings(input_str1)
print("Example 1 Output:", output1)  # Output: 7

# Example 2
input_str2 = "aba"
output2 = count_palindromic_substrings(input_str2)
print("Example 2 Output:", output2)  # Output: 4
```
